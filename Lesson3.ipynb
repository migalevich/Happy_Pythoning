```python
# Поиск лояльных клиентов
```

```python
user_df = df[['tc', 'art_sp']] # новый dataframe
```

```python
user_df
```

```python
user_df = user_df.rename(columns = {'tc':'user_id', 'art_sp':'brand_info'})
```

```python
user_df
```

```python
def split_brand(brand_name_data): 
    return brand_name_data.split(' ')[-1] 
```

```python
user_df['brand_name'] = user_df.brand_info.apply(split_brand) ## новая колонка 
```

```python
# user_df.brand_info.apply(lambda x: x.split(' ')[-1] )  # lambda функция
# более компактный способ получить название бренда
```

```python
user_df.head()
```

```python
users_purchases = user_df.groupby('user_id', as_index=False)\
    .agg({'brand_name':'count'})\
    .rename(columns = {'brand_name':'purchases'})\
    .query('purchases >= 5')
```

```python
users_purchases.purchases.median() ## среднее число покупок
```

```python
users_purchases.describe()
```

```python
users_unique_brands = user_df.groupby('user_id', as_index=False)\
    .agg({'brand_name':pd.Series.nunique})\
    .rename(columns = {'brand_name':'unique_brands'})
```

```python
lovely_brand_purchases_df = user_df.groupby(['user_id','brand_name'], as_index=False)\
    .agg({'brand_info':'count'})\
    .sort_values(['user_id','brand_info'], ascending=[False, False])\
    .groupby(['user_id'])\
    .head(1) \
    .rename(columns = {'brand_info':'lovely_brand_purchases', 'brand_name':'lovely_brand'})
```

```python
# объединение фреймов
loyalty_df = users_purchases\
    .merge(users_unique_brands, on = 'user_id')\
    .merge(lovely_brand_purchases_df, on = 'user_id')
```

```python
loyal_users = loyalty_df[loyalty_df.unique_brands == 1] # совершают много покупок, но одного бренда
```

```python
loyal_users 
```

```python
loyalty_df
```

```python
loyalty_df['loyalty_score'] = loyalty_df.lovely_brand_purchases / loyalty_df.purchases
```

```python
loyalty_df.loyalty_score.describe()
```

```python
ax = sns.displot(loyalty_df.loyalty_score) # график 
```

```python
loyalty_df.loyalty_score.median()
```

```python
brand_loyalty = loyalty_df.groupby('lovely_brand', as_index = False)\
    .agg({'loyalty_score':'median', 'user_id':'count'})\
    .sort_values(['loyalty_score'], ascending=[False])
```

```python
brand_loyalty
```

```python
ax = sns.barplot(x= 'lovely_brand', y='loyalty_score', data = brand_loyalty) # график 
```
